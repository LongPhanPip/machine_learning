```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
plt.plot(X, y, 'rx', label='Linear Regression')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Label', fontsize=15)
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]
X_b
```

```python
from numpy.linalg import inv
theta = inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta
```

```python
x_t = np.array([0, 2])
X_t = np.c_[np.ones((2, 1)), x_t]
y_t = X_t.dot(theta)
plt.figure(figsize=(12,10))
plt.plot(X, y, 'rx', label='Data')
plt.plot(x_t, y_t, c='darkorange', label='Line')
plt.legend(loc='best', fontsize=15)
plt.show()
```

```python
def sigmod(t) :
    return 1 / (np.exp(-t) + 1)
```

```python
step = 0.0001
m = 100
theta = np.random.randn(3, 1)
theta
```

```python
for i in range(100000):
    eta = X.T.dot(sigmod(X.dot(theta)) - y)
    theta -= (step * eta) / m
```

```python
theta
```

```python
x_t = np.array([-4, 4]).reshape(2, 1)
y_t = -theta[1] / theta[2] * x_t
y_t
```

```python
plt.figure(figsize=(12, 10))
plt.plot(x_t, y_t, c='y')
plt.plot(X1_r, X2_r, 'rx', label='Red')
plt.plot(X1_b, X2_b, 'b+', label='blue')
plt.legend(loc='best', fontsize=15)
plt.show()
```

```python
X = np.linspace(-10, 10, 1000)
y = sigmod(X)
```

```python
plt.figure(figsize=(12,10))
plt.title(r'$sigmod(x)$ = $\frac{1}{1 + e^{-x}}$', fontsize=20)
plt.plot(np.array([-10, 10]), np.array([1, 1]), c='r')
plt.plot(np.array([-10, 10]), np.array([0, 0]), c='r')
plt.plot(X, y, label=sigmod, c='orange')
```

```python
def softmax(X):
    X_exp = np.exp(X.T)
    return X_exp / X_exp.sum()
```

```python
theta = np.random.randn(3,3)
step = 0.005
for i in range(10000):
    gradient = X.T.dot(y - softmax(X.dot(theta)).T)
    theta = theta + step * gradient / 300
    
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.models.Sequential()
```

```python
model.add(keras.layers.Input(shape=400))
model.add(keras.layers.Dense(200, activation='relu'))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
import pandas as pd
X = pd.read_csv('data.csv', index_col=0).to_numpy() / 255
y = pd.read_csv('target.csv', index_col=0).to_numpy()
```

```python
model.compile(optimizer='SGD', loss='sparse_categorical_crossentropy')
```

```python
history = model.fit(X[:10000], y[:10000], epochs=30, validation_data=(X[10000:], y[10000:]))
```
