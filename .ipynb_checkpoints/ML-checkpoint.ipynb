```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
plt.plot(X, y, 'rx', label='Linear Regression')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Label', fontsize=15)
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]
```

```python
from numpy.linalg import inv
theta = inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta
```

```python
x_t = np.array([0, 2])
X_t = np.c_[np.ones((2, 1)), x_t]
y_t = X_t.dot(theta)
plt.figure(figsize=(12,10))
plt.plot(X, y, 'rx', label='Data')
plt.plot(x_t, y_t, c='darkorange', label='Line')
plt.legend(loc='best', fontsize=15)
plt.show()
```

```python
X1_r = np.random.randn(100, 1) - 1
X2_r = np.random.randn(100, 1) - 1
X1_b = np.random.randn(100, 1) + 1
X2_b = np.random.randn(100, 1) + 1
X_r = np.c_[np.ones((100, 1)), X1_r, X2_r]
X_b = np.c_[np.ones((100, 1)), X1_b, X2_b]
y_r = np.zeros((100, 1))
y_b = np.ones((100, 1))
y = np.r_[y_r, y_b]
X = np.r_[X_r, X_b]
```

```python
plt.figure(figsize=(12, 10))
plt.plot(X1_r, X2_r, 'rx', label='Red')
plt.plot(X1_b, X2_b, 'b+', label='blue')
plt.legend(loc='best', fontsize=15)
plt.show()
```

```python
def sigmod(t) :
    return 1 / (np.exp(-t) + 1)
```

```python
step = 0.0001
m = 100
theta = np.random.randn(3, 1)
theta
```

```python
for i in range(100000):
    eta = X.T.dot(sigmod(X.dot(theta)) - y)
    theta -= (step * eta) / m
```

```python
theta
```

```python
x_t = np.array([-4, 4]).reshape(2, 1)
y_t = -theta[1] / theta[2] * x_t
y_t
```

```python
plt.figure(figsize=(12, 10))
plt.plot(x_t, y_t, c='y')
plt.plot(X1_r, X2_r, 'rx', label='Red')
plt.plot(X1_b, X2_b, 'b+', label='blue')
plt.legend(loc='best', fontsize=15)
plt.show()
```

```python
X = np.linspace(-10, 10, 1000)
y = sigmod(X)
```

```python
plt.figure(figsize=(12,10))
plt.title(r'$sigmod(x)$ = $\frac{1}{1 + e^{-x}}$', fontsize=20)
plt.plot(np.array([-10, 10]), np.array([1, 1]), c='r')
plt.plot(np.array([-10, 10]), np.array([0, 0]), c='r')
plt.plot(X, y, label=sigmod, c='orange')
```

```python
X = 6 * np.random.rand(100, 1) - 4
y = 0.5 * X**2 + X + 2 + np.random.randn(100, 1)
```

```python
X1_r = np.random.randn(100) - 2
X1_b = np.random.randn(100)
X1_g = np.random.randn(100) + 2
X2_r = np.random.randn(100) + 2
X2_b = np.random.randn(100) - 2
X2_g = np.random.randn(100) + 2
X_r = np.c_[np.ones((100, 1)), X1_r, X2_r]
X_b = np.c_[np.ones((100, 1)), X1_b, X2_b]
X_g = np.c_[np.ones((100, 1)), X1_g, X2_g]
X = np.r_[X_r, X_b, X_g]
y_r = np.c_[np.ones((100, 1)), np.zeros((100, 1)), np.zeros((100))]
y_b =  np.c_[np.zeros((100)), np.ones((100, 1)), np.zeros((100))]
y_g = np.c_[np.zeros((100)), np.zeros((100)), np.ones((100, 1))]
y = np.r_[y_r, y_b, y_g]
```

```python
plt.figure(figsize=(12, 10))
plt.plot(X1_r, X2_r, 'rx', label='Red')
plt.plot(X1_b, X2_b, 'bx', label='Blue')
plt.plot(X1_g, X2_g, 'gx', label='Green')
plt.legend(loc='best', fontsize=16)
```

```python
def softmax(X):
    X_exp = np.exp(X.T)
    return X_exp / X_exp.sum()
```

```python
theta = np.random.randn(3,3)
step = 0.005
for i in range(10000):
    gradient = X.T.dot(y - softmax(X.dot(theta)).T)
    theta = theta + step * gradient / 300
```

```python
theta
```

```python
x1, x2 = np.meshgrid(
        np.linspace(-5, 5, 2000).reshape(-1, 1),
        np.linspace(-5, 5, 2000).reshape(-1, 1),
    )

x_t = np.c_[np.ones((x1.size, 1)), x1.ravel(), x2.ravel()]
```

```python
y_t = np.argmax(softmax(x_t.dot(theta)), axis=0)
y_t
```

```python
from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#d43b3b','#346bd1','#34d154'])
plt.figure(figsize=(12, 10))
plt.plot(X1_r, X2_r, 'rx', label='Red')
plt.plot(X1_b, X2_b, 'bx', label='Blue')
plt.plot(X1_g, X2_g, 'gx', label='Green')
plt.contourf(x1, x2, y_t.reshape(x1.shape), cmap=custom_cmap, alpha=0.5)
plt.legend(loc='best', fontsize=16, frameon=True, facecolor='w')
```
