```python
def preprocess(pos_path, neg_path, n_batch=32, n_shuffle_buffer=0, n_read_threads=5):
    p_dataset = tf.data.Dataset.list_files([os.path.join(pos_path, file) for file in os.listdir(pos_path)])
    p_dataset = dataset.interleave(lambda file: tf.data.TextLineDataset(file), cycle_length=5, num_parallel_calls=n_read_threads)
    p_dataset = dataset.map(lambda X: (X, 0), num_parallel_calls=n_read_threads)
    n_dataset = tf.data.Dataset.list_files([os.path.join(neg_path, file) for file in os.listdir(neg_path)])
    n_dataset = dataset.interleave(lambda file: tf.data.TextLineDataset(file), cycle_length=5, num_parallel_calls=n_read_threads)
    n_dataset = dataset.map(lambda X: (X, 1), num_parallel_calls=n_read_threads)
    dataset = tf.data.Dataset.concatenate(p_dataset, n_dataset).repeat(n_shuffle_buffer)
    return dataset.batch(n_batch).prefetch(1)
```

```python
test_pos_path = os.path.join(root, 'test/pos')
val_pos_path = os.path.join(root, 'val/pos')
test_neg_path = os.path.join(root, 'test/neg')
val_neg_path = os.path.join(root, 'val/neg')
train_pos_path = os.path.join(root, 'train/pos')
train_neg_path = os.path.join(root, 'val/neg')

n_batch = 32
n_shuffer_buffer = 25000 
train_set = preprocess(train_pos_path, train_neg_path, n_batch, n_shuffer_buffer)
test_set = preprocess(test_pos_path, test_neg_path, n_batch)
val_set = preprocess(val_pos_path, val_neg_path, n_batch)
```
